```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
baseball = pd.read_csv('baseball_data.csv')
sns.set(style="whitegrid")
baseball.describe()
```

```python
subset= baseball[baseball['avg']>0]
len(subset)
subset.describe()
```

```python
majority = subset[subset['height']<=77]
```

```python
ax = sns.countplot(baseball['height'], color ='lightblue')
ax.set_title('Distribution of Players Height ')
```

```python
ax = sns.regplot(x='height', y='weight',data=baseball, color="g")
ax.set_title('Weight vs. Height ')
```

```python
ax = sns.violinplot(x="height", y="avg", data=baseball, palette="Set3")
ax.set_title('Violinplot - Avg vs.Handedness ')
```

```python
ax = sns.boxplot(x="handedness", y="HR", data=baseball, palette="Set3")
```

```python
ax = sns.distplot(baseball['avg'], color ='lightblue')
```
